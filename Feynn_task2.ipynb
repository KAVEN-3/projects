```python
sb.countplot(x= df['PowerTrain'])
plt.title('Count Plot of Powertrain', fontsize = 20)
plt.xlabel('Power Train', fontsize = 15)
plt.ylabel('Count', fontsize = 15)
```

```python
ax= plt.figure(figsize=(15,8))
sb.heatmap(df.corr(),linewidths=1,linecolor='white',annot=True)
```

```python
a=np.arange(1,103)

ax= plt.figure(figsize=(20,5))
sb.barplot(x='Brand',y=a,data=df,palette='ch:s=-.2,r=.6')
plt.grid(axis='y')
plt.title('Brands in the datset')
plt.xlabel('Brand')
plt.ylabel('PriceEuro')
plt.xticks(rotation=90)
```

```python
a=np.arange(1,103)

ax= plt.figure(figsize=(20,5))
sb.barplot(x='Brand',y=a,data=df,palette='ch:start=.2,rot=-.3')
plt.grid(axis='y')
plt.title('Brands in the datset')
plt.xlabel('Brand')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
```

```python
ax= plt.figure(figsize=(20,5))
sb.barplot(x='Brand',y='Range_Km',data=df,palette='crest')
plt.grid(axis='y')
plt.title('Maximum Range achieved by a brand')
plt.xlabel('Brand')
plt.ylabel('Range')
plt.xticks(rotation=90)
```

```python
ax= plt.figure(figsize=(20,5))
sb.barplot(x='Brand',y='Efficiency_WhKm',data=df,palette='ch:s=-.2,r=.6')
plt.grid(axis='y')
plt.title('Efficiency achieved by a brand')
plt.xlabel('Brand')
plt.ylabel('Efficiency')
plt.xticks(rotation=90)
```

```python
ax= plt.figure(figsize=(20,5))
sb.barplot(x='Brand',y='PriceEuro',data=df,palette='crest')
plt.title('Price of a Car')
plt.xlabel('Price in Euro')
plt.grid(axis='y')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
```
